```python
import glob
import os

def list_all_excel_files(path, recursive=False):
    
    return_value = list()
    
    if recursive is False: 
        for filename in glob.glob(os.path.join(path,"*.*")):
            if filename.lower().endswith(".xls") or filename.lower().endswith(".xlsx"):
                return_value.append(filename)
    else:
        for filename in glob.glob(os.path.join(path,"**","*.*"), recursive=True):
            if filename.lower().endswith(".xls") or filename.lower().endswith(".xlsx"):
                return_value.append(filename)
    return return_value


search_path = r"D:\Well Engineering\Mapping\EXCEL per Folder\FWR"

import pprint
paths = list_all_excel_files(path=search_path, recursive = True)
paths = [os.path.basename(p) for p in paths]


pprint.pprint(paths)
```
